```python
import pickle
file_path = "/home/ssawmya-local/FM_work/SmartEM/smartem/segmentation/exclusion_area_per_section.pkl"
exclusion_area = pickle.load(open(file_path, "rb"))

all_exclusions = []
for section in exclusion_area:
    all_exclusions.extend(exclusion_area[section])


print("Total number of exclusions: ", len(all_exclusions))
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.hist(np.array(all_exclusions)*100, bins = 100)
plt.xlabel("% Exclusion")
plt.ylabel("Frequency")

# draw a line for the mean
mean_exclusion = sum(all_exclusions) / len(all_exclusions)
mean_exclusion = mean_exclusion * 100
plt.axvline(mean_exclusion, color='k', linestyle='dashed', linewidth=1)

# write the mean value beside the line
plt.text(mean_exclusion+2, 800, 'Mean: {:.2f} %'.format(mean_exclusion), color='blue', fontsize=10)

# plt.show()
plt.savefig("exclusion_area_histogram.png")
```
